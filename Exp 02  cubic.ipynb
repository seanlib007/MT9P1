```python
def cub(xd,xp):
    y=spline(xd)
    n=len(xd)
    print(y(xp))
    print(y.derivative(xp))
    plt=point(xd,color="grey",size=20)
    plt=plt+plot(y,(x,xd[0][0],xd[-1][0]),color="red")
    yprime=lambda x:y.derivative(x)
    plt=plt+plot(yprime,(x,xd[0][0],xd[-1][0]),color="green")
    show(plt)
```

```python
cub([(1,-8),(2,-1),(3,18)],2.5)
cub([(0,0),(pi/4,1/sqrt(2)),(pi/2,1),(3*pi/4,1/sqrt(2)),(pi,0)],pi/6)
```
